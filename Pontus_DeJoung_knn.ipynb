```python
import numpy as np

def analyze_dataset(df):
    # Grundläggande statistik
    print("Grundläggande statistik för datasetet:")
    print(df.describe())

    # Kontrollera saknade värden
    print("\nAntal saknade värden i datasetet:")
    print(df.isnull().sum().sum())

    # Histogram över pixelvärden
    print("\nHistogram över pixelvärden:")
    pixel_values = df.iloc[:, 1:].values.flatten()
    plt.hist(pixel_values, bins=20, color='gray')
    plt.title('Pixelvärden')
    plt.show()

    # Klassfördelning
    print("\nKlassfördelning:")
    plt.hist(df['label'], bins=np.arange(11)-0.5, color='gray', rwidth=0.8)
    plt.title('Klassfördelning')
    plt.xticks(np.arange(10))
    plt.show()

    # Boxplot för att detektera outliers
    print("\nBoxplot för att detektera outliers:")
    plt.boxplot(pixel_values)
    plt.title('Boxplot över pixelvärden')
    plt.show()
```

```python
# Analysera de två träningsdataseten
print("Analys av träningsdata 1:")
analyze_dataset(train_df1)
print("Analys av testsdata 1:")
analyze_dataset(test_df1)
print("\nAnalys av träningsdata 2:")
analyze_dataset(train_df2)
print("Analys av testdata 2:")
analyze_dataset(test_df2)
```

# Träna en KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split

X_train = train_df1.drop('label', axis=1).values
y_train = train_df1['label'].values
X_test = test_df1.drop('label', axis=1).values
y_test = test_df1['label'].values

k_values = [1, 3, 5, 7, 9, 11]
best_k = k_values[0]
best_accuracy = 0
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k) 
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    if accuracy > best_accuracy:
        best_k = k
        best_accuracy = accuracy

    print(f'k={k}, accuracy={accuracy}')
    
print(f'Best k={best_k}, with accuracy={best_accuracy}')
knn_best = KNeighborsClassifier(n_neighbors=best_k)
knn_best.fit(X_train, y_train)
```

# Skapa en confusion matrix för att se hur modellen klassificerar

```python
import seaborn as sns

y_pred = knn_best.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
labels = np.unique(y_test)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)

plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```

# ○ Tänk på tidskomplexiteten (finns inget krav på en viss tidskomplexitet)

```python
import numpy as np
from scipy.spatial import distance
from collections import Counter

class MyKNN:
    def __init__(self, n_neighbors=3):
        self.n_neighbors = n_neighbors

    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

    def predict(self, X):
        predicted_labels = [self._predict(x) for x in X]
        return np.array(predicted_labels)

    def _predict(self, x):
        distances = distance.cdist([x], self.X_train, 'euclidean')[0]
        k_indices = distances.argsort()[:self.n_neighbors]
        k_nearest_labels = self.y_train[k_indices]
        most_common = Counter(k_nearest_labels).most_common(1)
        return most_common[0][0]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

X_train = train_df1.drop('label', axis=1).values
y_train = train_df1['label'].values
X_test = test_df1.drop('label', axis=1).values
y_test = test_df1['label']

knn = MyKNN(n_neighbors=3)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print(classification_report(y_test, y_pred))
accuracy = accuracy_score(y_test, y_pred)
print("Test Score:",accuracy)
```
